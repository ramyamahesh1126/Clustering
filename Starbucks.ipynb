```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
#Data Set
X = np.array([[1,5],
    [1,7],
    [2,6],
    [2,9],
    [3,6],
    [3,8],
    [3,9],
    [4,8],
    [3,3],
    [5,4],
    [7,2],
    [5,7],
    [4,5],
    [6,9],
    [7,3],
    [7,9],
    [8,1],
    [8,7]])

labels = range(1, 19)
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()

#3 clusters
ac2 = AgglomerativeClustering(n_clusters = 3,affinity = 'euclidean',linkage='ward')
y_hc=ac2.fit_predict(X)
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], s=100, c='red', label ='Cluster 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], s=100, c='blue', label ='Cluster 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], s=100, c='green', label ='Cluster 3')
plt.title('Clusters of Starbucks (Hierarchical Clustering Model)')
plt.show()
  
# # Visualizing the clustering
# plt.figure(figsize =(6, 6))
# plt.scatter(X[0], X[1], 
#            c = ac2.fit_predict(X, y=None), cmap ='rainbow')
# plt.show()

#Single_linkage
linked = linkage(X, 'single')

labelList = range(1, 19)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()

#Complete linkage
linked = linkage(X, 'complete')

labelList = range(1, 19)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()

#Average linkage
linked = linkage(X, 'average')

labelList = range(1, 19)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```
